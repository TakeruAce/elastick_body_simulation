```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
theta = np.array([0.0,0.0,0.0])
theta_prev = np.array([0.0,0.0,0.0])
omega = np.array([0.0,0.0,0.0])
x = np.array([0.0,0.0,0.0])
m1 = 0.1 # kg
m2 = 0.05 # kg
l1 = 0.1 # m
l2 = 0.1 # m
c = 0.02 # 
k = 0.5
dt = 0.002 # 500 Hz
sim_time = 5
Ff = np.array([0.0,0.0,0.0])

thetas = np.array([])
omegas = np.array([])
Ffs = np.array([])
```

```python
def update(th,th_prev, Ff_):
    th[2] = - k*th_prev[0] / (m2*l2*l2) - c * th_prev[1] / (m2*l2) - (x[2] + (l1+l2)* omega[1]) / l2
    th[1] = th[1] + th[2] * dt
    th[0] = th[0] + th[1] * dt

    Ff_[0] = k * th[0] / l2
    th_prev[0] = th[0]
    th_prev[1] = th[1]
```

```python
thetas = np.array([])
omegas = np.array([])
Ffs = np.array([])
t_sim = np.arange(0,sim_time,dt)
for t in t_sim:
    # if t < 0.5:
    #     omega[1] = 100 * np.sin(2*np.pi*2*t)
    # else:
    #     omega[1] = 0
    
    update(th=theta, th_prev=theta_prev,Ff_=Ff)
    thetas = np.append(thetas, theta[0])
    omegas = np.append(omegas, omega[1])
    Ffs = np.append(Ffs, Ff[0])
```
